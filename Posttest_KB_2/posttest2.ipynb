#### Ketentuan 2: 

- Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. 
- Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
# init
countRecord = len(df.index)
listAttributes = list(df.columns)
countDataAttributeObj = 0
countDataAttributeInt = 0
countDataAttributeWithNull = 0

# loop mencari atribut mengandung null
for item in list(df.isnull().any()):
  if(item == True):
    countDataAttributeWithNull+=1

# loop mencari atribut tipe angka atau objek
for item in df.dtypes:
  if (item == 'object'):
    countDataAttributeObj+=1
  elif (item == 'int64' or item == 'float64'):
    countDataAttributeInt+=1
```

```python
# dictionary
dataAttr = {
  'jumlah_record': countRecord,
  'nama_attribute': listAttributes,
  'attribute_objek': countDataAttributeObj,
  'attribute_angka': countDataAttributeInt,
  'attribute_mengandung_null': countDataAttributeWithNull
}

print("{")
print(" Jumlah Record: {data}".format(
  data = dataAttr["jumlah_record"]
))
print(" Nama Attribute: {data}".format(
  data = dataAttr["nama_attribute"]
))
print(" Attribute object: {data}".format(
  data = dataAttr["attribute_objek"]
))
print(" Attribute Angka: {data}".format(
  data = dataAttr["attribute_angka"]
))
print(" Attribute mengandung null: {data}".format(
  data = dataAttr["attribute_mengandung_null"]
))
print("}")
```

#### Ketentuan 3: 

Tampilkan 
 - rata-rata,

 - nilai minimum,
