Import Pandas & NumPy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employee_data.csv')
df.head()
```

```python
data_pie = df['department'].value_counts().rename_axis('Department').reset_index(name='Id_count')
plt.figure(figsize=(10, 12))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.Department, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Department')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Dari data yang ditampilkan pada diagram di atas, dapat terlihat bahwa department yang terbanyak adalah sales, disusul oleh engineering, dll.

```python
plt.figure(figsize = (10, 8))
Attrplot=sns.countplot(x = 'status', data = df, palette='Set2')
plt.title("status")
plt.show()


#observations
#people who leave are usually less satisfied of their job(role)
```

Dari diagram yang terlihat di atas, terlihat bahwa ada sekitar 3800 karyawan yang memutuskan untuk meninggalkan pekerjaannya, sisanya memilih untuk tetap bekerja

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='status',data=df, hue='department', palette='Set2')
plt.title('status vs department')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Terlihat pada diagram di atas, data pekerja yang masih bekerja dan data pekerja yang meninggalkan pekerjaannya
yang paling banyak meninggalkan pekerjaannya adalah sales dan engineering

```python
sns.boxplot(x='status',data=df,y='satisfaction', palette='Set2')
plt.title('Status vs Satisfaction')
#observations
#people who leave are usually less satisfied of their job(role)
```

Setelah dilakukan observasi, dapat terlihat bahwa penyebab para pekerja yang keluar dari pekerjaannya adalah karena ketidakpuasannya terhadap pekerjaannya.

Data yang ditampilkan di bawah ini menjelaskan mengapa para pekerja tidak puas atas pekerjaannya

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='salary',data=df, hue='department', palette='Set2')
plt.title('salary vs department')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('Jumlah')
plt.xlabel('Gaji')
plt.show()
```

Terlihat bahwa gaji yang diterima para pekerja kebanyakan dalam skala low dan medium
